# Examples of One Dimensional plotting

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
np.random.seed(1000)
y = np.random.standard_normal(20)
```

We create sample indeces to plot y data.

```python
x = range(len(y))
plt.plot(x,y)
```

However, that's not required. You can pass just y values and pyplot will be smart enought

```python
plt.plot(y.cumsum())
```
